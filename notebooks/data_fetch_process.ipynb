```python
import yfinance as yf
import pandas as pd
import wbdata as wbd


economy_tickers = {"NY.GDP.MKTP.CD": "GDP (current US$)",
                       "NY.GDP.MKTP.KD.ZG": "GDP growth rate (%)",
                       "SL.UEM.TOTL.ZS": "Unemployment rate (%)",
                       "FR.INR.RINR": "Real interest rate (%)",
                       "NY.GDP.DEFL.KD.ZG": "Inflation rate (%)",
                       "SP.POP.TOTL": "Population",
                       "NE.EXP.GNFS.CD": "Exports (current US$)",
                       "NE.IMP.GNFS.CD": "Imports (current US$)",
                       }
start_date = "2001-01-01"
end_date = "2024-01-01"
interval = "1d"
country="USA"

economics_data = wbd.get_dataframe(
            economy_tickers, country, date=(start_date[:4], end_date[:4]))


economics_data = economics_data.sort_index(ascending=True)


economics_data["Population growth (%)"] = economics_data["Population"].pct_change(fill_method=None) * 100

economics_data["Export growth (%)"] = economics_data["Exports (current US$)"].pct_change(fill_method=None) * 100

economics_data["Import growth (%)"] = economics_data["Imports (current US$)"].pct_change(fill_method=None) * 100





economics_data.head(20)



```
