```python
import yfinance as yf
import pandas as pd

df = yf.download("^GSPC", start="2001-01-01", end="2024-01-01")


# Calculate price change percentage
df['PriceChange%'] = (df['Close'] - df['Open']) / df['Open'] * 100

# Add a moving average of volume (7-day example, adjust for your needs)
df['AvgVolume'] = df['Volume'].rolling(window=7, min_periods=1).mean()

# Calculate volume ratio
df[('VolumeChange%', '')] = (df[('Volume', '^GSPC')] - df[('AvgVolume', '')]) / df[('AvgVolume', '')] * 100

# Drop rows with missing values
df.dropna(inplace=True)




# Calculate Sentiment score - Price change more important
df['SentimentScore'] = df['PriceChange%'] * 3 + df['VolumeChange%']

# Determine sentiment label
df['Sentiment'] = pd.cut(df['SentimentScore'], 
    bins=[-float('inf'), -5, 8, float('inf')], 
    labels=['Bearish', 'Neutral', 'Bullish'])

df.sample(20)






```
